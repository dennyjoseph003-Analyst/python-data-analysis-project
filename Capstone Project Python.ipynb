```python
#import pandas for data frame operations

import pandas as pd

#import numpy for numerical operations

import numpy as np

#TASK 1

# create project data frame

project_data = {"ID":["A001","A002","A003","A004","A005","A002","A005","A003","A001","A003","A001","A004","A004","A005"],
    "Project":["Project 1","Project 2","Project 3","Project 4","Project 5","Project 6","Project 7","Project 8","Project 9","Project 10","Project 11","Project 12","Project 13","Project 14"],
    "Cost":[1002000,2000000,4500000,5500000,None,680000,400000,350000,None,300000,2000000,1000000,3000000,200000],
    "Status":["Finished","Ongoing","Finished","Ongoing","Finished","Failed","Finished","Failed","Ongoing","Finished","Failed","Ongoing","Finished","Finished"]}
project_df = pd.DataFrame(project_data)

# save to csv

project_df.to_csv("project.csv", index = False)

project_df
```

```python
# create employee data frame

employee_data = {"ID":["A001","A002","A003","A004","A005"],
                 "Name":["John Alter","Alice Luxumberg","Tom Sabestine","Nina Adgra","Amy Johny"],
                 "Gender":["M","F","M","F","F"],
                 "City":["Paris","London","Berlin","Newyork","Madrid"],
                 "Age":[25,27,29,31,30]}  
employee_df = pd.DataFrame(employee_data)

# save to csv

employee_df.to_csv("employee.csv", index = False)

employee_df
```

```python
# Create seniority level data frame

seniority_level_data = {"ID":["A001","A002","A003","A004","A005"],
                        "Designation Level":[2,2,3,2,3]}
seniority_df = pd.DataFrame(seniority_level_data)

# save to csv

seniority_df.to_csv("seniority.csv", index = False)

seniority_df
```

```python
# TASK 2

# Load Project DataFrame

project_df = pd.read_csv("project.csv")

# convert cost column to list of iteration

costs = project_df["Cost"].tolist()

# loop through costs

for i in range(len(costs)):
    if costs[i] is None or (isinstance(costs[i], float) and np.isnan(costs[i])):
        prev_values = [c for c in costs[:i] if c is not None]
        costs[i] = sum(prev_values)/len(prev_values)
project_df.loc[:,"Cost"]=costs

project_df        
```

```python
# TASK 3

# Load employee data frame

employee = pd.read_csv("employee.csv")

# split name column

employee_df[["First name","Last name"]] = employee_df["Name"].str.split(" ", n=1, expand=True)

# Drop old name column

employee_df = employee_df.drop(columns=["Name"])

# desired order

employee_df = employee_df[["ID","First name", "Last name","Gender","City","Age"]]

employee_df
```

```python
# TASK 4

# load seniority dataframe

seniority_df = pd.read_csv("seniority.csv")

# merge project and employee dataframe

merged = pd.merge(project_df,employee_df, on="ID")

# meerge with seniority dataframe

Final = pd.merge(merged,seniority_df, on="ID")

Final
```

```python
# Task 5

# Add Bonus column: 5% of Cost if Status is Finished, else 0

Final["Bonus"] = np.where(Final["Status"]=="Finished", Final["Cost"]*0.05,0)

Final
```

```python
# Task 6

# Demote designation level by 1 if project failed

Final.loc[Final["Status"] == "Failed", "Designation Level"]+=1

# Drop employees whose designation level > 4 (ineligible to lead projects)

Final = Final[Final["Designation Level"]<=4]

Final
```

```python
# Task 7

# Add Mr. or Mrs. prefix based on Gender

Final["First name"] = np.where(Final["Gender"]=="M","Mr. "+Final["First name"], "Mrs. "+Final["First name"])

# Drop Gender column since prefix replaces it

Final.drop(columns=["Gender"], inplace=True)

Final
```

```python
# Task 8

# Promote designation level by 1 for the employees age > 29

for i in range (len(Final)):
    if Final.loc[i,"Age"]>29:
        Final.loc[i,"Designation Level"]-=1

Final
```

```python
# Task 9

# group by ID and first name, sum project costs

total_proj_cost = Final.groupby(["ID","First name"])["Cost"].sum().reset_index()

# rename columns cost to total cost

total_proj_cost.rename(columns={"Cost":"Total Cost"}, inplace=True)

# save as csv

total_proj_cost.to_csv("TotalProjCost.csv", index=False)

total_proj_cost
```
